```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def rossler(t, Y, omega, a, b, c):
    x, y, z = Y
    dx_dt = -omega * y - z
    dy_dt = omega * x + a * y
    dz_dt = b + z * (x - c)
    return [dx_dt, dy_dt, dz_dt]
```

```python
a = 0.16
b = 0.1
c = 8.5
omega = 1
```

```python
sol = solve_ivp(rossler, [2000, 2300], [5, 5, 0], args=(omega, a, b, c), dense_output=True)
```

```python
t = np.linspace(2000, 2300, 300)
res = sol.sol(t)    # dense_output=True
```

```python
plt.xlim(2000, 2300)
plt.ylim(-20, 20)
plt.xlabel('t')
plt.ylabel('x')
plt.plot(t, res[0])
```

```python
plt.xlim(2000, 2300)
plt.ylim(-20, 20)
plt.xlabel('t')
plt.ylabel('y')
plt.plot(t, res[1])
```

```python
plt.xlim(2000, 2300)
plt.xlabel('t')
plt.ylabel('z')
plt.plot(t, res[2])
```
